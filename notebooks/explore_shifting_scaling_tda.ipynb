```python
# Create artificial data, corresponding to randomly rotated circle in 3D
theta = np.linspace(0, 2 * np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)
z = np.zeros_like(x)

# Randomly rotate circle
theta = np.random.uniform(0, 2 * np.pi)
rot = np.array([[np.cos(theta), -np.sin(theta), 0],
                [np.sin(theta), np.cos(theta), 0],
                [0, 0, 1]])
data = np.dot(rot, np.array([x, y, z]))

# Plot data in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(data[0], data[1], data[2])
plt.show()

```
